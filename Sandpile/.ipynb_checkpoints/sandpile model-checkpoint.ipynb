```python
def dropsand(grid,dim):
    grid[dim//2,dim//2]=grid[dim//2,dim//2]+1
    return grid

def stabilitycheck(grid,dim,z_c):
    t=1
    topplings=[]                          #topplings after adding a single grain of sand
    while(t!=0):
        unstable=[]
        for i in range(1,dim+1):
            for j in range(1,dim+1):
                if(grid[i,j]>=z_c):
                    unstable.append([i,j])
        t=len(unstable)
        topplings.append(t)
        for x in unstable:
            grid[x[0],x[1]]=grid[x[0],x[1]]-4
            grid[x[0]+1,x[1]]=grid[x[0]+1,x[1]]+1
            grid[x[0]-1,x[1]]=grid[x[0]-1,x[1]]+1
            grid[x[0],x[1]+1]=grid[x[0],x[1]+1]+1
            grid[x[0],x[1]-1]=grid[x[0],x[1]-1]+1
        #
        grid[:,0]=0
        grid[0,:]=0
        grid[:,dim+1]=0
        grid[dim+1,:]=0
    
    return grid,topplings

#len(topplings) is the duration of avalanche.
```

```python
import numpy as np
from sympy.interactive import printing
import matplotlib.pyplot as plt
printing.init_printing(use_latex= True)

dim=20
z_c=4
grid_1d=np.zeros((dim + 2)*(dim + 2), dtype=float)
for i in range((dim+2)**2):
    grid_1d[i]=int(3)  #np.random.random()*4
grid=np.reshape(grid_1d,(dim+2,dim+2))

#Boundary_Conditions
grid[:,0]=0
grid[0,:]=0
grid[:,dim+1]=0
grid[dim+1,:]=0

print(grid)
plt.imshow(grid)
plt.colorbar()
```

```python
a,b=stabilitycheck(dropsand(grid,dim),dim,z_c)
print(a,b)
plt.imshow(a)
plt.colorbar()
```
